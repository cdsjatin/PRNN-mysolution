```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stat

%matplotlib inline
```

```python
## 0 --- 1 = 0 pos
## 1 -- -1 = 1 neg

train = pd.read_csv('./datasets_assignement1/P1c_test_data.txt',skiprows=0, header=None)
    
print(len(train))
```

```python
X_train = train.values[:,0:2]
y_train = train.values[:,2]

print(X_train.shape, y_train.shape)

X_train_0 = X_train[:200]
y_train_0 = y_train[:200]

X_train_1 = X_train[200:]
y_train_1 = y_train[200:]

print("Train shape: ", X_train_0.shape, X_train_1.shape)
print("Labels shpae: ", y_train_0.shape, y_train_1.shape)
```

```python
mu0 = np.average(X_train_0, axis=0)
cov0 = np.cov(X_train_0.T)

mu1 = np.average(X_train_1, axis=0)
cov1 = np.cov(X_train_1.T)

print(mu0)
print(cov0)
print(mu1)
print(cov1)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
a = plt.hist2d(X_train_0[:,0], X_train_0[:,1])
plt.colorbar()

plt.subplot(1,2,2)
a = plt.hist2d(X_train_1[:,0],X_train_1[:,1], alpha=0.8)

plt.colorbar()
```

```python
# Bayes Classifier
test = pd.read_csv('./datasets_assignement1/P1c_test_data.txt',skiprows=0, header=None)
    
print(len(test))
```

```python
X_test = test.values[:,0:2]
y_test = test.values[:,2]

print(X_test.shape, y_test.shape)
```

```python
def accuracy(y_predict, y_test):
    return np.average(y_predict == y_test)
```

```python
def P(x, Class=None):
    if Class == 0:
        return stat.multivariate_normal(mu0, cov0).pdf(x)
    else:
        return stat.multivariate_normal(mu1, cov1).pdf(x)
```

```python
y_predict = [] ## 0 --- 1 = 0 pos
               ## 1 -- -1 = 1 neg
for i in range(len(test)):
    if P(X_test[i,:], 0) > P(X_test[i,:], 1):
        y_predict.append(1)
    else:
        y_predict.append(-1)
        
print(accuracy(y_predict, y_test))
```
